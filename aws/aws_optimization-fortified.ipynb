```python
datestrings = df['datetime'].to_list()
dates = [dateutil.parser.parse(s) for s in datestrings]

fig = plt.figure(figsize=(12,6))
ax=plt.gca()

ax.set_title("AWS EC2 Utilizations", fontsize=20)
ax.set_ylabel('CPU Utilization Percentages', color='black',fontsize=18)
ax.set_xlabel('Date', color='black',fontsize=18)

dateticks = [20200301,20200308,20200315,20200322,20200329]
dateticksX = [dt.datetime.strptime(str(int(dtx)),'%Y%m%d') for dtx in dateticks]

ax.set_xticks(dateticksX)
xfmt = mdates.DateFormatter('%m-%d')
ax.xaxis.set_major_formatter(xfmt)
plt.plot(dates,df['i-089389afc8097dac4'], color="blue")
plt.plot(dates,df['i-09f0182499b5eaf27'], color="red")
plt.plot(dates,df['i-0ca364ec6656ff517'], color="green")
plt.legend(['i-089389afc8097dac4','i-09f0182499b5eaf27','i-0ca364ec6656ff517'], fontsize=18)
plt.show()
```
